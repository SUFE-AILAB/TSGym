```python
from itertools import product
```

```python
# gym_series_norm=None,
# gym_series_decomp=None,
# gym_input_embed='series-patching',
# gym_network_architecture='MLP',
# gym_attn='sparse-attention',
# gym_encoder_only=True,
```

```python
model_names = ['_'.join(list(_)) for _ in list(product(['TSGym'],
                                                       ['False', 'True'], # gym_series_sampling
                                                       ['None', 'Stat', 'RevIN', 'DishTS'], # gym_series_norm
                                                       ['None', 'MA', 'MoEMA', 'DFT'], # gym_series_decomp
                                                       ['False', 'True'], # gym_channel_independent
                                                       ['series-encoding', 'series-patching'], # gym_input_embed
                                                       ['Transformer'], # gym_network_architecture
                                                       ['self-attention', 'auto-correlation', 'sparse-attention', 'frequency-enhanced-attention'], # gym_attn
                                                       ['True'] # gym_encoder_only
                                                       ))] +\
                                                       ['_'.join(list(_)) for _ in list(product(['TSGym'],
                                                       ['False', 'True'], # gym_series_sampling
                                                       ['None', 'Stat', 'RevIN', 'DishTS'], # gym_series_norm
                                                       ['None', 'MA', 'MoEMA', 'DFT'], # gym_series_decomp
                                                       ['False', 'True'], # gym_channel_independent
                                                       ['series-encoding', 'series-patching'], # gym_input_embed
                                                       ['MLP', 'GRU'], # gym_network_architecture
                                                       ['null'], # gym_attn
                                                       ['True'] # gym_encoder_only
                                                       ))]
```

```python
model_names = [_ for _ in model_names if 'False_series-patching' not in _]
```

```python
print(len(model_names), model_names)
```

```python
# 模板文件路径
template_path = '/data/coding/chaochuan/TSGym/scripts/long_term_forecast/ECL_script/debug.sh'
# 输出目录
output_dir = '/data/coding/chaochuan/TSGym/scripts/long_term_forecast/ECL_script/gym'

# 确保输出目录存在
import os
os.makedirs(output_dir, exist_ok=True)

# 读取模板内容
with open(template_path, 'r') as file:
    template_content = file.read()

# 对于每个模型名称，生成一个 shell 脚本
for model_name in model_names:
    # 替换模型名称
    script_content = template_content.replace('TransformerGym', model_name)
    
    # 定义输出文件名
    output_file = os.path.join(output_dir, f'{model_name}.sh')
    
    # 写入新的 shell 脚本
    with open(output_file, 'w') as file:
        file.write(script_content)

    print(f'Generated {output_file}')
```
